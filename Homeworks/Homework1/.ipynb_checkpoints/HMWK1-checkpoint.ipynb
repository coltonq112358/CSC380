# Problem 2: Pie Chart

```python
starbucks_df["Beverage"].value_counts()
```

```python
import matplotlib.pyplot as plt

beverage_category_counts = starbucks_df["Beverage"].value_counts()
labels = beverage_category_counts.keys()
# sizes = 

fig, ax = plt.subplots()
ax.pie(beverage_category_counts, labels=labels)
plt.show()
```
